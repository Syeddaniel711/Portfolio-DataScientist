```python
# Find the optimal number of clusters using the elbow method
wcss = []
max_clusters = 20  # You can adjust the maximum number of clusters to consider

for num_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(df_selected_pos)
    wcss.append(kmeans.inertia_)  # Inertia: Sum of squared distances to the closest centroid

# Step 5: Plot the Elbow Plot
plt.plot(range(1, max_clusters + 1), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Plot')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 1: Load the dataset
data = pd.read_csv("Asignment4.csv", sep=';', encoding='latin-1')

# Step 2: Feature Selection
selected_features = data.drop(["Rk", "Nation", "Age", "Born", "Squad", "Player", "Comp"], axis=1)

selected_pos = ["FW", "FWMF", "MF", "MFFW"]
df_selected_pos = selected_features[selected_features["Pos"].isin(selected_pos)].drop(columns=["Pos"])

# Step 3: Remove any rows with missing values (optional but can help with the analysis)
df_selected_pos.dropna(inplace=True)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_selected_pos)

# Step 4: Cluster Analysis using k-means
num_clusters = 5  # You can experiment with different values here
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)

# Step 5: Graphical Display
# Plotting the clusters based on the 'Pos' variable and two performance metrics (you can choose different metrics)
metric1 = 'Goals'
metric2 = 'Assists'

plt.figure(figsize=(10, 6))
for cluster in range(num_clusters):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data[metric1], cluster_data[metric2], label=f'Cluster {cluster}')

plt.xlabel(metric1)
plt.ylabel(metric2)
plt.title('Cluster Analysis of Football Players')
plt.legend()
plt.grid(True)
plt.show()

```

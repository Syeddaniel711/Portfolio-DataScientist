```python
Football=pd.read_csv("Asignment4.csv",sep=';', encoding='latin-1')
pd.set_option('display.max_columns', None) 
Football
```

```python
Football=Football.drop(["Rk","Nation","Age","Born","Squad","Comp"],axis=1)

# Handle missing values if any
Football=Football.dropna()
Football.describe()
```

```python
Defensive = ['DF', 'DFFW', "FWDF"]
Data1 = Football[Football["Pos"].isin(Defensive)].drop(columns=["Pos"])
Data1_new=Data1[[
    'Tkl', 'TklWon', 'TklDef3rd', 'TklMid3rd', 'TklAtt3rd', 'TklDri', 'TklDriAtt', 'TklDri%',
    'TklDriPast', 'Blocks', 'BlkSh', 'BlkPass', 'Int', 'Tkl+Int', 'Clr', 'Recov',
    'AerWon', 'AerLost', 'AerWon%','Fls', 'CarMis', 'CarDis','Player'
]]

# Setting "Player" as the index
Data1_new.set_index("Player", inplace=True)
Data1_new
```

```python
from sklearn.preprocessing import scale
X = pd.DataFrame(scale(Data1_new), index=Data1_new.index, columns=Data1_new.columns)
with pd.option_context('display.max_rows', None):
   print(X.dtypes)
X
```

```python
from sklearn.decomposition import PCA
pca_loadings = pd.DataFrame(PCA(n_components=2).fit(X).components_.T, index=Data1_new.columns,columns=["V1","V2"])
pca_loadings

# Fit the PCA model and transform X to get the principal components
pca = PCA(n_components=2)
df_plot = pd.DataFrame(pca.fit_transform(X), index=X.index,columns=["PC1","PC2"])
df_plot
```

```python
df_plot
```

```python
PC1 = pca.fit_transform(X)[:,0]
PC2 = pca.fit_transform(X)[:,1]
ldngs = pca.components_
scalePC1 = 1.0/(PC1.max() - PC1.min())
scalePC2 = 1.0/(PC2.max() - PC2.min())
```

```python
from sklearn.cluster import KMeans
# Perform K-Means clustering on the principal components (PC1 and PC2)
kmeans = KMeans(n_clusters=5, n_init='auto')  # Choose the number of clusters you want
clusters = kmeans.fit_predict(df_plot)

# Add a new column 'Cluster' to df_plot to store the cluster labels
df_plot['Cluster'] = clusters
plt.figure(figsize=(14,9))
features = Data1_new.columns
for i, feature in enumerate(features):
    plt.arrow(0, 0, ldngs[0, i], 
             ldngs[1, i], 
              head_width=0.03, 
             head_length=0.03)
    plt.text(ldngs[0, i] * 1.15, 
            ldngs[1, i] * 1.15, 
            feature, fontsize=18)
 
sns.scatterplot(x=PC1 * scalePC1,
                y=PC2 * scalePC2, 
                hue='Cluster',
                data=df_plot,
                palette="viridis")
 
plt.xlabel('PC1', fontsize=20)
plt.ylabel('PC2', fontsize=20)
plt.title('K-Means Clustering on PCA Components For defensive position', fontsize=20)
```

```python
# Find the optimal number of clusters using the elbow method
wcss = []
max_clusters = 20  # You can adjust the maximum number of clusters to consider

for num_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(Data1_New)
    wcss.append(kmeans.inertia_)  # Inertia: Sum of squared distances to the closest centroid

# Step 5: Plot the Elbow Plot
plt.plot(range(1, max_clusters + 1), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Defensive position for Elbow Plot')
plt.show()
```
